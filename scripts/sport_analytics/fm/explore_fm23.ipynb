```python
pos_cols = ['GK', 'DL', 'DC', 'DR', 'WBL',
            'WBR', 'DM', 'ML', 'MC', 'MR',
            'AML', 'AMC', 'AMR', 'ST']

def max_position(row):
    return row[pos_cols].idxmax()

df['main_position'] = (
    df[pos_cols].idxmax(axis=1)
)
```

Zentrale Positionen sind fast immer Rechts

```python
df["main_position_centered"] = [i.replace("R","_outside").replace("L","_outside") for i in df["main_position"]]
```

```python
grouped_df  = df.groupby(['main_position_centered', 'right_footed'])[attributes].mean().round(2)
grouped_df
```

```python
pivot = (
    df
    .groupby(['main_position_centered', 'right_footed'])[attributes]
    .mean()
    .unstack('right_footed')
)

left  = pivot.xs(False, level='right_footed', axis=1)
right = pivot.xs(True,  level='right_footed', axis=1)

pct_lr = ((left - right) / right).round(2)
pct_lr
```

# Technical

## Technical Attributes – Left vs Right (Left − Right)

### Overall Pattern
- Differences are small but more structured than physical attributes.
- Clear tendency: **Left-footed players score slightly higher in offensive technical skills.**
- Right-footed players show small advantages in central defensive attributes.

---

## Offensive / Creative Skills

### Crossing
- Positive across nearly all roles.
- Strongest in DC (+0.13), GK (+0.12), AMC/MC (+0.07).

→ Most consistent technical advantage for left-footed players.

---

### Long Shots
- Positive in almost all positions.
- Strongest in:
  - GK (+0.24)
  - D_outside (+0.10)
  - WB_outside (+0.10)

→ Reinforces earlier set-piece / shooting tendency.

---

### Dribbling & Technique
- Small positives in AMC, MC, DC.
- Large GK values likely noise (+0.20).
- No major structural effect.

---

### Finishing
- Slight positives in AMC.
- Slight negatives in DC, DM, ST.
- No consistent striker effect.

---

## Defensive Attributes

### Marking
- Slight negative trend in central roles:
  - DM (−0.07)
  - MC (−0.04)

→ Small right-foot defensive bias centrally.

---

### Tackling
- Negative in MC (−0.06), DM (−0.04).
- Positive in M_outside (+0.06), ST (+0.05).

→ Defensive central players slightly right-foot dominant.

---

### Heading
- Slight negative in DM (−0.06).
- Otherwise near zero.

---

## Passing
- Small positives in most roles.
- No structural dominance.

---

## GK
- Very large positive technical differences:
  - Dribbling (+0.20)
  - Long Shots (+0.24)
  - Heading (+0.19)

Likely sampling noise, not meaningful.

---

## Conclusion
- Left-footed players show small but consistent advantages in:
  - Crossing
  - Long Shots
  - Offensive technical attributes

- Right-footed players slightly stronger in:
  - Marking
  - Tackling
  - Central defensive robustness

Effect sizes remain small, but technical patterns are more consistent than physical ones.

# Physical

# Blog Structure – Left vs Right Footed Players (FM Data Analysis)

## 1. Hook / Opening Idea

- Are left-footed players actually different — or just rarer?
- Does footedness shape player profiles, or only tactical usage?
- A data-driven look at positional, technical, mental, and physical differences.

---

## 2. Core Finding (High-Level)

- Differences exist, but are small overall.
- Technical attributes show the clearest pattern.
- Mental and physical differences are minimal.
- Footedness appears more role-driven than biologically driven.

---

## 3. Technical Profile – Where It Matters

### Strongest Pattern

Left-footed players show higher averages in:
- Crossing  
- Long Shots  
- Set-piece related skills  

### Positional Nuance

- Wide roles (Fullbacks, Wingbacks, Wide Midfielders) show clearer left-foot technical bias.
- Defensive Midfielders show a slight negative technical trend.
- GK technical spike is likely a sampling effect.

### Interpretation Angle

- Coaches may assign left-footed players to creative or set-piece roles.
- Scarcity may increase specialization.

---

## 4. Mental Profile – Subtle Defensive Bias

Right-footed players are slightly higher in:
- Aggression  
- Positioning  
- Defensive mentality (especially DM, MC)

- Differences are small but consistent in central roles.
- No strong intelligence or creativity bias.

### Angle for the Blog

- Central defensive roles appear slightly right-foot dominated in mentality.
- Possible cultural or tactical filtering effect.

---

## 5. Physical Profile – Almost No Difference

- Acceleration, Pace, Stamina, Strength: near zero difference.
- No structural athletic advantage for either side.
- Confirms that footedness ≠ physical archetype.

**Strong statement:**  
Footedness is not athletic destiny.

---

## 6. Positional Insights

### Most Interesting Positions

- Defensive Midfielder → most negative mental/technical bias.
- Wide roles → strongest technical positivity.
- Goalkeeper → statistical outlier.

### Narrative Direction

- Footedness interacts with tactical deployment.
- Selection bias is likely stronger than innate skill bias.

---

## 7. Big Takeaways

- The myth of the “magical left foot” is exaggerated.
- The only consistent difference: slight technical/set-piece orientation.
- Mental and physical profiles remain largely neutral.
- Coaching and role assignment likely explain more than biology.
