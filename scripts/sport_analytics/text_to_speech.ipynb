# Change Voice

```python
from transformers import AutoProcessor, BarkModel

processor = AutoProcessor.from_pretrained("suno/bark")
model = BarkModel.from_pretrained("suno/bark")

voice_preset = "v2/en_speaker_6"
voice_preset = "v2/en_speaker_1"

```

```python
from IPython.display import Audio
text_prompt = """Welcome to Statsfaction! Sports is a wild and unpredictable world, but in contrast to your toxic ex-girlfriend, sports will never abandon you. The same goes for math. We've all had those nightmares of forgotten math homework, being called up to the front, and sweating like Dennis Rodman in front of an ATM. But there's more to it than meets the eye.
Utilizing statistics effectively can deepen your understanding of your favorite sport. It allows you to rediscover your love for your favorite players and build a compelling case against your rivals.
Today, I embark on a journey to explore the potential of AI in sports. I'm eager to uncover hidden secrets and determine which will remain elusive. Join me as we delve into the intersection of AI and sports, unraveling the possibilities and unearthing untapped potential. """
inputs = processor(text_prompt, voice_preset=voice_preset)

audio_array = model.generate(**inputs)
audio_array = audio_array.cpu().numpy().squeeze()

sample_rate = model.generation_config.sample_rate
Audio(audio_array, rate=sample_rate)
```

```python
import scipy.io.wavfile as wavfile
wavfile.write("test.wav", rate=sample_rate, data=audio_array)
```
